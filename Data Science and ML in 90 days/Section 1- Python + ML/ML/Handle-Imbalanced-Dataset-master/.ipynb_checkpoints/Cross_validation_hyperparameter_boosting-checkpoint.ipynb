```python
import pandas as pd
data = pd.read_csv("creditcard.csv")
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
X = data.drop("Class",axis=1)
y = data.Class
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,train_size=0.7)
```

# Cross validation using kFold and Hyperparameter tuning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
logistic_class = LogisticRegression()
grid = {'penalty':['l1','l2'],'C':10.0**np.arange(-2,3),'random_state':[0,50,100,200]}
cv = KFold(n_splits = 5,random_state=[0,50,100,200],shuffle=False)
```

```python
classifier = GridSearchCV(logistic_class,grid,cv=cv,n_jobs=-1,scoring="roc_auc")
classifier.fit(X_train,y_train)
```

```python
print(classifier.best_params_)
print(classifier.best_score_)
```

```python
y_pred = classifier.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
