```python
import random

#Global variables
suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':11, 'Queen':12, 'King':13, 'Ace':14}
```

```python
class Card:
    '''
    Class to initialize cards with suit and rank
    '''
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        self.value = values[rank]
        
    def __str__(self):
        return self.suit + 'of' + self.rank
```

```python
class Deck:
    '''
    Class to create a deck of Cards. Eg: King of hearts, queen of spades
    '''
    def __init__(self):
        self.whole_deckofcards = [] # An empty list to store all objects created of class Card
        
        for suit in suits:
            for rank in ranks:
                object_created = Card(suit,rank) # Creating object of Card
                
                self.whole_deckofcards.append(object_created) # Appending the object to the emty list.
                
                #                  OR another method
                #self.deckofcards.append(Card(suit,rank))
    
    def shuffle(self):
        random.shuffle(self.whole_deckofcards) #Randomly shuffling the deck. Note: random.shuffle doesn't return anything
        
    def deal_one_card(self):
        return self.whole_deckofcards.pop()   #This method deals one card whenever called
```

```python
class Player:
    def __init__(self,name):
        self.name = name
        
        self.player_deckofcards = []
        
    def remove_cards(self):
        return self.player_deckofcards.pop(0)
        
    def add_cards(self,new_cards):
        if type(new_cards)==type([]):
            self.player_deckofcards.extend(new_cards)
        else:
            self.player_deckofcards.append(new_cards)
    def __str__(self):
        return f'Player {self.name} has {len(self.player_deckofcards)} cards '
```

```python
#Game logic

player_one = Player("Rajdeep")
player_two = Player("Pratik")

new_deck = Deck()
new_deck.shuffle()

for i in range(26):
    player_one.add_cards(new_deck.deal_one_card())
    player_two.add_cards(new_deck.deal_one_card())
```

```python
game_on = True
```

```python
round_num = 0

while game_on:
    round_num+=1
    print(f'Round no:{round_num}')
    
    if len(player_one.player_deckofcards)==0:
        print("Player one has no cards left")
        print("Player two wins!!")
        game_on = False
        break
        
    if len(player_two.player_deckofcards)==0:
        print("Player two has no cards left")
        print("Player one wins!!")
        game_on = False
        break
        
    player_one_hand_cards = []
    player_one_hand_cards.append(player_one.remove_cards())
    
    player_two_hand_cards = []
    player_two_hand_cards.append(player_two.remove_cards())
    
    war_on = True
    
    while war_on:
        if player_one_hand_cards[-1].value > player_two_hand_cards[-1].value:
            player_one.add_cards(player_one_hand_cards)
            player_one.add_cards(player_two_hand_cards)
            
            war_on = False
            
        elif player_one_hand_cards[-1].value < player_two_hand_cards[-1].value:
            player_two.add_cards(player_two_hand_cards)
            player_two.add_cards(player_one_hand_cards)
            
            war_on = False
            
        else:
            print("WAAAAAAR!!!!!!")
            
            if len(player_one.player_deckofcards)<5:
                print("Player one does not have enough cards for war.")
                print("Player Two wins!")
                game_on=False
                break
            
            elif len(player_two.player_deckofcards)<5:
                print("Player Two does not have enough cards for war.")
                print("Player One wins!")
                game_on=False
                break
            else:
                for i in range(5):
                    player_one_hand_cards.append(player_one.remove_cards())
                    player_two_hand_cards.append(player_two.remove_cards())
```
